```python
import mira
import anndata
import pandas as pd
```

```python
atac_data = anndata.read_h5ad('mira-datasets/shareseq_Ma2020/atac_data.joint_representation.motif_calls.tss_annotated.h5ad')
atac_model = mira.topics.AccessibilityTopicModel.load('mira-datasets/shareseq_Ma2020/atac_model.pth')
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
order = dendrogram(atac_data.uns['topic_dendogram'])['leaves']
```

```python
topic_activations = pd.DataFrame(atac_data[order, atac_model.features].varm['topic_feature_compositions'])
```

```python
for col in ['end','start','chr']:
    topic_activations.insert(0, col, atac_data.var[col].values)
```

```python
topic_activations\
    .sort_values(['chr','start'])\
    .to_csv('data/trackdata/topics_matrix.bedgraph', index = None, header = None, sep = '\t')
```
